# Source separation

This notebook shows how to use the functions developed for souce separation.

## 1. Imports

First we import the library of functions.

```python
from utilities_store import *
```

## 2. Obtain spectrogram of the original sources and the mixture

First we obtain the spectrogram of each individual source and the mixture of the sources.
We also obtain the dynamic resonances for representation purposes.

#### Source 1

```python
spectrogram_1, signal_1 = get_spectrogram('nine_mono_20kHz.wav', N = 300, step_size = 150, 
                                          power_threshold = 1e-9, amp_threshold = 1e-8)
```

```python
dyns_1 = get_dynamic_resonances(spectrogram_1, min_overlap = 1e-15, overlap_type = "exponential", 
                                freq_ratio = 15, mode = 'forward', multiple_match = False)
```

#### Source 2

```python
spectrogram_2, signal_2 = get_spectrogram('anybody-here_20kHz.wav', N = 300, step_size = 150, 
                                          power_threshold = 1e-9, amp_threshold = 1e-8)
```

```python
dyns_2 = get_dynamic_resonances(spectrogram_2, min_overlap = 1e-15, overlap_type = "exponential", 
                                freq_ratio = 15, mode = 'forward', multiple_match = False)
```

#### Mixture

```python
spectrogram_3, signal_3 = get_spectrogram('anybody-here_nine_20kHz.wav', N = 600, step_size = 200, 
                                          power_threshold = 1e-9, amp_threshold = 1e-8)
```

```python
dyns_3 = get_dynamic_resonances(spectrogram_3, min_overlap = 1e-15, overlap_type = "exponential", 
                                freq_ratio = 15, mode = 'forward', multiple_match = False)
```

Plot the Dynamic resonance spectrogram of the two sources and the mixture. We also add in the background the Fast Fourier spcetrogram of the mixture:

```python
plot_dynamic_overlapped(spectrogram_1, dyns_1, spectrogram_2, dyns_2, spectrogram_3, signal_3, dyns_3, 
                            min_freq = 0, max_freq = 2000, plot_fourier = True, scale = 1)
```

## 3. Obtain the fundamental frequency (f0) of each source from the mixture

We obtain the fundamental frequency of each of the sources using only the mixture. Multiple fundamental frequencies(two or more sources) are obtained by finding maximum harmonicity resonances that are not integer multiple of each other (harmonic between themselves). After these maximum harmonicity resonances are found, we match each resonance with the corresponding source.

```python
f0_s = get_f0(spectrogram_3, ratio = 0.1, n_sources = 2)
```

Plot the fundamental frequency (f0) of each source over the spectrogram of the mixture:

```python
plot_f0(spectrogram_3, f0_s, min_freq = 0, max_freq = 1000)
```

## 4. Separate the sources

Next we separate each source from the mixture based on the fundamental frequency. Each resonance of the spectrogram is added to a source attenuated by a factor. This factor depends on how far this resonance is in frequency to the closest integer multiple of the fundamental frequency of the corresponding source.

The attenuation applied to each resonance is calculated as:

<br>
<center>$\normalsize Distance = | \: ( \: Resonance frequency \: \% \: f0 \: ) \: / \: f0 \: | $</center>
<br>
<center>$\normalsize Attenuation = e ^{- \: Distance \: \cdot  \: Attenuation \: factor}  $</center>

Where % denotes modulo operation

```python
source_0, source_1 = source_separation(spectrogram_3, f0_s, max_dist = 0.05, att_factor = 20)
```

Next we plot the resonance spectrogram of each separated source

#### Source 0

```python
plot_spectrogram(source_0, max_freq = 2000, scale=500)
```

#### Source 1

```python
plot_spectrogram(source_1, max_freq = 2000, scale=500)
```

Finally, we reproduce the audio of the mixture and each separated source:

#### Mixture

```python
ipd.Audio(data = signal_3, rate = spectrogram_3.sample_rate[0])
```

#### Source 0

```python
ipd.Audio(data = source_0.reconstruction.real, rate = source_0.sample_rate[0])
```

#### Source 1

```python
ipd.Audio(data = source_1.reconstruction.real, rate = source_1.sample_rate[0])
```
